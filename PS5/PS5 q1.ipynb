```python
import numpy as np
import matplotlib.pyplot as plt
```

Define function f and g, the rates of Notch and Delta activation, respectively

```python
def f(D):
    return ((D**2)/(0.1 + D**2))
```

```python
def g(N):
    return (1/(1 + 10*N**2))
```

Define the dynamical equations of Delta activation

```python
def D1Rate(D1,D2):
    return (g(f(D2)) - D1)
```

```python
def D2Rate(D1,D2):
    return(g(f(D1)) - D2)
```

Plot the (D1Rate,D2Rate) vector field in the D1,D2 plane

```python
w = 1
Y, X = np.mgrid[0:w:100j, 0:w:1000j]

fig = plt.figure(figsize = (12, 7))
plt.streamplot(X, Y, D1Rate(X,Y), D2Rate(X,Y), density = 1)
plt.xlabel('D1');
plt.ylabel('D2');
plt.title('Streamplot of (D1Rate,D2Rate)')
plt.show()
```

Calculate the nullclines, D2 vs D1 for D1Rate and D2Rate = 0. 

The two nullclines are inverses of each other, so solving for D2 in one of the D/dTau equations and plotting it alongside its inverse yields the nullcline plot.

```python
import sympy
from sympy.solvers import solve
from sympy import Symbol

D1, D2 = sympy.symbols('D1,D2')
sympy.solve(D2Rate(D1,D2),D2)
```

```python
def D2(D1):
    return 100.0*(D1**2 + 0.1)**2/(1000.0*D1**4 + 100.0*(D1**2 + 0.1)**2)
```

Plot the nullclines and observe intersection points

```python
D1 = np.linspace(0,1)
plt.plot(D2(D1),D1)
plt.plot(D1,D2(D1))
plt.legend(['D2one','D2two'])
plt.xlabel('D1');
plt.ylabel('D2');
plt.title('Nullcline curves yielded for D1Rate = 0, D2Rate = 0')
```

The streamplot and nullclines demonstrate that three steady-state solutions exist: high delta 1 and low delta 2, low delta one and high delta 2, or equal delta 1 and delta 2. This is similar to the result seen in the limit of delta degradation much greater than notch degradation, although the delta concentrations do not reach their maximum values in the limit of this problem.
